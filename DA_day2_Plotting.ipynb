Problem 2: Plot the scores of 2 students in 5 different subjects .
Subjects as x-axis and  marks as y-axis

```python
stu1 = [56,78,92,95,78]
stu2 = [77,89,91,23,45]
sub = ["Maths","Science","English","Telugu","Social"]
plt.plot(sub,stu1,label="student1")
plt.plot(sub,stu2,label="student2")
plt.title("marks vs subject")
plt.legend()
```

```python
stu1 = [56,78,92,95,78]
stu2 = [77,89,91,23,45]
sub = ["Maths","Science","English","Telugu","Social"]
plt.subplot(2,1,1)
plt.plot(sub,stu1,label="student1",color="m")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,stu2,label="student2",color="y")
plt.legend()
```

Problem 3:
Create Subplots(bar) showing the profits of two companies.
Given : Revenue of company A and B in lakhs
plt bar()
year=(2019,2020,2021,2022,2023)
A=(230,560,780,127,128)
B=(200,160,270,127,400)

```python
A = np.array([230,560,780,127,128])
B = np.array([200,160,270,127,400])
print("Revenues of Company A " ,A)
print("Revenues of Company B " ,B)
profA = np.diff(A)
print("Profits of A " ,profA)
profB = np.diff(B)
print("Profits of B " ,profB)
X = ["19-20","20-21","21-22","22-23"]
plt.subplot(1,2,1)
plt.bar(X,profA,color='k' ,label = "profit A")
plt.legend(loc = "best")
plt.subplot(1,2,2)
plt.bar(X,profB,color='g',label = "profit B")
plt.legend(loc = "best")
```

```python
a = np.array([25,60,5,10])
labe = ["AIML","Python","Pandas","Numpy"]
plt.pie(a,labels = labe)
plt.show()
```

```python
#explode
a = np.array([25,60,5,10])
labe = ["AIML","Python","Pandas","Numpy"]
explo = [0.2,0,0,0]
plt.pie(a,labels = labe,explode = explo,startangle = 180)
plt.show()
```
